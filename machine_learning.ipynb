## Importação das bibliotecas utilizadas

```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
import sklearn.model_selection as ms
import sklearn.linear_model as lm
```

```python
db = pd.read_csv(r"D:\Users\JOSUEH HENRIQUE\Downloads\desafio_indicium_imdb.csv")
```

```python
db.head(1)
```

Aqui faremos um tratamento mais rigoroso, não queremos treinar a máquina com dados duvidosos, então excluiremos os nulos.

```python
db = db.dropna()
db.isnull().sum()
db.nunique()
db['Gross'] = db['Gross'].str.replace(',', '').astype(float)
db['Runtime'] = db['Runtime'].str.replace('min', '').astype(int)
```

## Prevendo a nota do IMDB
Começaremos separando os dados que julgamos importantes e interessante para o machine learning.

Esses dados serão transformados em dummies para o funcionamento do treinamento da máquina.

```python
x = db.iloc[:, [2, 3, 4, 5, 8, 9, 10, 11, 12, 13, 14, 15]]
y = db.iloc[:, 6]

x_dum = pd.get_dummies(x)
```

Separaremos os espaços para treinamento e testes da máquina.
E enfim o treinamento e teste com a regressão linear.

```python
x_train, x_test, y_train, y_test = ms.train_test_split(x_dum, y, test_size = 1/5, random_state = 0)
```

```python
regressor = lm.LinearRegression()
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)

np.set_printoptions(precision = 2)
y_concat = np.concatenate((y_pred.values.reshape(len(y_pred),1), y_test.values.reshape(len(y_test),1)),1)
```

Resultado:

```python
x_test = x_test.reset_index(drop = True)

y_compare = pd.DataFrame(y_concat)
y_compare = y_compare.rename(index = str, columns = {0 : 'y_pred', 1 : 'y_test'})
y_compare = y_compare.reset_index(drop = True)

final = pd.concat([y_compare, x_test], axis = 1)
```
